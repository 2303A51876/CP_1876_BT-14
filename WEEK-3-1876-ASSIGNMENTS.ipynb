```python
#T.Shylasri-2303A51876(BT-14)
#WEEK-3 ASSIGNMENTS
#Friday Lab(23-1-26)
#1Q) Assignment 1: N-Queens Problem
def solveNQueens(n):
    col = set()
    diag1 = set()
    diag2 = set()
    board = [["."] * n for _ in range(n)]
    def backtrack(r):
        if r == n:
            for row in board:
                print(" ".join(row))
            print()
            return
        for c in range(n):
            if c in col or (r - c) in diag1 or (r + c) in diag2:
                continue
            col.add(c)
            diag1.add(r - c)
            diag2.add(r + c)
            board[r][c] = "Q"
            backtrack(r + 1)
            col.remove(c)
            diag1.remove(r - c)
            diag2.remove(r + c)
            board[r][c] = "."
    backtrack(0)
n = int(input())
solveNQueens(n)
```

```python
#T.Shylasri-2303A51876(BT-14)
#WEEK-3 ASSIGNMENTS
#Friday Lab(23-1-26)
#2Q) Assignment 2: LCS â€“ Longest Common Subsequence
X = input()
Y = input()
m, n = len(X), len(Y)
dp = [[0]*(n+1) for _ in range(m+1)]
for i in range(1, m+1):
    for j in range(1, n+1):
        if X[i-1] == Y[j-1]:
            dp[i][j] = 1 + dp[i-1][j-1]
        else:
            dp[i][j] = max(dp[i-1][j], dp[i][j-1])
print(dp[m][n])
```

```python
#T.Shylasri-2303A51876(BT-14)
#WEEK-3 ASSIGNMENTS
#Friday Lab(23-1-26)
#3Q) Assignment 3: Sum of Substrings
s = input()
n = len(s)
total = 0
for i in range(n):
    num = 0
    for j in range(i, n):
        num = num * 10 + int(s[j])
        total += num
print(total)
```
