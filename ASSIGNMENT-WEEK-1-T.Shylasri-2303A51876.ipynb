```python
#ASSIGNMENT-1
#WEDNESDAY LAB(07-01-2026)
#2303A51876-T.Shylasri(BATCH-14)
t = int(input())
for _ in range(t):
    n = int(input())
    jobs = []

    for _ in range(n):
        d, p = map(int, input().split())
        jobs.append((p, d))
    # Sort jobs by profit descending
    jobs.sort(reverse=True)
    max_deadline = max(d for _, d in jobs)
    slots = [-1] * (max_deadline + 1)
    jobs_done = 0
    total_profit = 0
    for profit, deadline in jobs:
        for t in range(deadline, 0, -1):
            if slots[t] == -1:
                slots[t] = profit
                jobs_done += 1
                total_profit += profit
                break
    print(jobs_done, total_profit)
```

```python
#ASSIGNMENT-2
#WEDNESDAY LAB(07-01-2026)
#2303A51876-T.Shylasri(BATCH-14)
def max_crossing_sum(arr, left, mid, right):
    left_sum = float('-inf')
    total = 0
    for i in range(mid, left - 1, -1):
        total += arr[i]
        left_sum = max(left_sum, total)
    right_sum = float('-inf')
    total = 0
    for i in range(mid + 1, right + 1):
        total += arr[i]
        right_sum = max(right_sum, total)

    return left_sum + right_sum
def max_subarray(arr, left, right):
    if left == right:
        return arr[left]

    mid = (left + right) // 2

    return max(
        max_subarray(arr, left, mid),
        max_subarray(arr, mid + 1, right),
        max_crossing_sum(arr, left, mid, right)
    )
t = int(input())
for _ in range(t):
    n = int(input())
    arr = list(map(int, input().split()))
    print(max_subarray(arr, 0, n - 1))
```

```python
#ASSIGNMENT-1
#FRIDAY LAB(09-01-2026)
#2303A51876-T.Shylasri(BATCH-14)
import heapq
t = int(input())

for _ in range(t):
    n = int(input())
    cables = list(map(int, input().split()))

    heapq.heapify(cables)
    total_cost = 0

    while len(cables) > 1:
        a = heapq.heappop(cables)
        b = heapq.heappop(cables)
        cost = a + b
        total_cost += cost
        heapq.heappush(cables, cost)

    print(total_cost)
```

```python
#ASSIGNMENT-2
#FRIDAY LAB(09-01-2026)
#2303A51876-T.Shylasri(BATCH-14)
def quickselect(arr, k):
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    if k <= len(left):
        return quickselect(left, k)
    elif k <= len(left) + len(mid):
        return pivot
    else:
        return quickselect(right, k - len(left) - len(mid))
t = int(input())
for _ in range(t):
    n, k = map(int, input().split())
    arr = list(map(int, input().split()))
    print(quickselect(arr, k))
```
